# Selecting rows,columns based on condition

import libraries

```python
import numpy as np
import pandas as pd
```

read csv file

```python
df = pd.read_csv('titanic_train.csv')
df.head(10)
```

to get passengers list who are majors

```python
df[df['Age'] >= 18]
```

to get list of female passengers

```python
df[df['Sex'] == 'female']
```

to know whose ticket fare is 500 and above

```python
df[df['Fare'] >= 500]
```

if you have multiple qureies use AND(&) , OR(|)

```python
df_q = df[(df['Age'] >= 18) & (df['Sex'] == 'female')]
print(df_q)
```
