# Data-Cleaning

Importing required libraries

```python
import pandas as pd
import numpy as np
```

Reading the dataset and printing first five rows

```python
data = pd.read_csv('titanic_train.csv')
data.head()
```

changing the index coloumn from default to PassengerId

```python
data = pd.read_csv('titanic_train.csv', index_col='PassengerId')
data.head(10)
```

To fill null data in coloumns only with int datatype with zeros 
and coloumns with string datatype with a any made default string here 'mid'

```python
data = data.fillna({'Age':0 ,'Cabin':'mid'}) 
data.head(10)
```

in some scenarios we may need to fill missing values in strings or int,with their next value or previous value
so, data = data.fillna(method = 'ffill') /fills values same as before row
data = data.fillna(method = 'bfill')   /fills values same as later row

data = data.interpolate() /all int col null values are replaced by avg of above and below values
let nth row be a nullvalue, now after interpolate() nth value will be
(n+1)th + (n-1)th / 2
