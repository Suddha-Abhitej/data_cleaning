# handling duplicates

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic_train.csv',index_col = 'PassengerId')
df.head(10)
```

to check if any two rows are similar 
first ensures only first simialr row is considered original and rest all are duplicate
none is default value we can replace with any col required name 

```python
df.duplicated(subset = None, keep = 'first')
```

```python
df.duplicated(subset = ['Ticket'], keep = 'first')
```

to print particular row

```python
df.loc[[889]]
```

drops duplicate tickets,and we can see the rows are reduced to 681

```python
df.drop_duplicates('Ticket',inplace=True)
df.duplicated('Ticket')
```
