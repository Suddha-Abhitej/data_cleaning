# functions in pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic_train.csv',index_col = 'PassengerId')
df.head(10)
```

creating afunction to sort age of passengers into child,adult,senior citizen

```python
def age_sort(x):
    if x < 18:
        return 'Child'
    if x>=18 and x<=45:
        return 'Adult'
    else:
        return 'Senior Citizen'
    
df['Age'] = df['Age'].apply(age_sort)
```

```python
df.head(10)
```

using seaborn to plot bargraph

```python
import seaborn as sns
sns.countplot(df['Age'])
```

to get information on how many passengers survived

```python
def survive_list(x):
    if x == 1:
        return 'Survived'
    else:
        return 'Dead'
    
df['Survived'] = df['Survived'].apply(survive_list)
```

```python
sns.countplot(df['Survived'])
```

to know how many female and male passengers are there

```python
def gender(x):
    if x == 'male':
        return 'male'
    else:
        return 'female'
    
df['Sex'] = df['Sex'].apply(gender)
```

```python
sns.countplot(df['Sex'])
```
